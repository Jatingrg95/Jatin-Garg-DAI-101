```python
# Step 1: Identify numerical columns (excluding 'id')
numerical_columns = data.select_dtypes(include=['number']).columns
numerical_columns = [col for col in numerical_columns if col.lower() != 'id']  # Exclude 'id'

# Step 2: Outlier Detection using IQR method
def detect_outliers_iqr(df, numerical_columns):
    outliers = {}
    for col in numerical_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

# Get outliers based on IQR method
numerical_outliers = detect_outliers_iqr(data, numerical_columns)

# Print outliers for numerical columns
print(f"Outliers detected in columns:")
for col, outlier_data in numerical_outliers.items():
    print(col)
    # Step 3: Create a Boxplot for all numerical columns (excluding 'id')
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[numerical_columns])  # Only numerical columns excluding 'id'
plt.title('Outliers in Numerial Columns', fontsize=16)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
from scipy.stats import chi2_contingency
# List of categorical columns for analysis
categorical_columns = ['Gender', 'City', 'Profession', 'Dietary Habits', 'Degree', 
                       'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']

# Gender vs Family History of Mental Illness
ct_gender_family_history = pd.crosstab(data['Gender'], data['Family History of Mental Illness'])
chi2_gender_family_history, p_gender_family_history, _, _ = chi2_contingency(ct_gender_family_history)


# Dietary Habits vs Sleep Duration
ct_dietary_sleep = pd.crosstab(data['Dietary Habits'], data['Sleep Duration'])
chi2_dietary_sleep, p_dietary_sleep, _, _ = chi2_contingency(ct_dietary_sleep)


# Profession vs Financial Stress (if Financial Stress is categorical, otherwise skip this step)
# Here, assuming 'Financial Stress' has been categorized (e.g., low, medium, high)

data['Financial Stress Category'] = pd.cut(data['Financial Stress'], bins=[-float('inf'), 3, 6, float('inf')], labels=['Low', 'Medium', 'High'])

ct_profession_financial_stress = pd.crosstab(data['Profession'], data['Financial Stress Category'])

chi2_profession_financial_stress, p_profession_financial_stress, _, _ = chi2_contingency(ct_profession_financial_stress)

# Output Chi-Square Results
print(f"\nChi-Square Test Results:")
print(f"Gender vs Family History of Mental Illness p-value: {p_gender_family_history}")
print(f"Dietary Habits vs Sleep Duration p-value: {p_dietary_sleep}")
print(f"Profession vs Financial Stress p-value: {p_profession_financial_stress}")
```

```python
# Create a subplot for visualizing all the cross-tabulations (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Gender vs Family History of Mental Illness
sns.heatmap(
    ct_gender_family_history, 
    annot=True, 
    fmt="d", 
    cmap="Blues", 
    ax=axes[0, 0], 
    linecolor='black', 
    linewidths=1
)
axes[0, 0].set_title(
    "Gender vs Family History of Mental Illness\nChi2 p-value: {:.3f}".format(p_gender_family_history),
    fontsize=14
)
axes[0, 0].tick_params(axis='x', colors='black', labelsize=12)
axes[0, 0].tick_params(axis='y', colors='black', labelsize=12)
# Plot 2: Dietary Habits vs Sleep Duration
sns.heatmap(
    ct_dietary_sleep, 
    annot=True, 
    fmt="d", 
    cmap="Blues", 
    ax=axes[0, 1], 
    linecolor='black', 
    linewidths=1
)
axes[0, 1].set_title(
    "Dietary Habits vs Sleep Duration\nChi2 p-value: {:.3f}".format(p_dietary_sleep),
    fontsize=14
)
axes[0, 1].tick_params(axis='x', colors='black', labelsize=12)
axes[0, 1].tick_params(axis='y', colors='black', labelsize=12)
# Plot 3: Profession vs Financial Stress
sns.heatmap(
    ct_profession_financial_stress, 
    annot=True, 
    fmt="d", 
    cmap="Blues", 
    ax=axes[1, 0], 
    linecolor='black', 
    linewidths=1
)
axes[1, 0].set_title(
    "Profession vs Financial Stress\nChi2 p-value: {:.3f}".format(p_profession_financial_stress),
    fontsize=14
)
axes[1, 0].tick_params(axis='x', colors='black', labelsize=12)
axes[1, 0].tick_params(axis='y', colors='black', labelsize=12)
# Remove the empty subplot (axes[1, 1])
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
"""
If all p-values are significant (< 0.05), then there is a strong relationship between the compared variables.
If p-values are greater than 0.05, theres no statistically significant association.
The heatmaps highlight which categories contribute most to observed patterns.
"""
```
